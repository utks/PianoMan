We start with a theoretical overview of angular momentum of light. Classical polarization optics mainly consisted of examining paraxial light and it's 2D polarization states. This implied that the spin angular momentum of light is directly related to their circular polarization, which in turn led to the conclusion that the spin of a totally unpolarized light vanishes and it's polarization is undefined. However later studies led to the discovery of the transverse spin components of a non-paraxial light which was orthogonal to the main propagation direction and was largely independent of the polarization state of the wave. For this we have to consider all three spatial components of the field vector and we also need to extend the existing polarization theory to the 3D case. Then we go on to demonstrate that the transverse spin survives even in non-paraxial fields generated from totally unpolarized paraxial light; the non-paraxial field being a focused beam and an evanescent wave in our case.

##**Background**
Angular  momentum  (AM)  was  recognized  as  one  of  the  important  characteristics  of light after the pioneering work by Poynting and the first experimental evidence by Beth. These works considered the spin AM produced by the circular polarization of a light beam. 

The  spin  AM is  associated with the polarization of  light,  so  that right-hand  and  left-hand circular  polarizations of a paraxial beam correspond  to  the  positive  and  negative helicities $\sigma=\pm1$ of  photons.

Now, in the majority of situations the spin angular momentum of light is  aligned  with  its mean momentum k (the familiar $\vec{E}.\vec{H}=0)$. In  other  words,  it  is longitudinal. Recently, there  has been a  rapidly  growing  interest in optical  fields  with  a transverse AM (both spin and orbital, we will focus on spin).  Such  angular momentum  is  orthogonal  to  the propagation  direction (mean momentum) of  light.

However, there is a more intriguing transverse spin AM which was  first  described  in  2012 by  Bliokh  and  Nori.  This  transverse  spin appears locally in structured optical fields,  such  as  evanescent  waves,  focused  beams,  and  two-wave  interference. It is independent of the helicity of the wave and can appear even for linearly polarized waves.

# **Spin and orbital angular momenta: Basic Properties**
**Point Particle**

---
The angular momentum of a particle is defined as:
$\vec{L}=\vec{r}\times\vec{p}$, 
where $\vec{r}$ and $\vec{p}$ are position and angular momentum of particle.
This angular momentum is extrinsic, i.e. dependent on the choice of origin of the coordinate system.

By definition,  a point  particle  cannot  have  any  *internal  structure*  and,  hence,  cannot carry an intrinsic AM. Evidently, the AM is transverse, i.e., orthogonal to the momentum.

**Plane Wave**

---
In contrast to localized particles, waves are extended entities. The limiting case of a wave entity is  a plane  wave.  Such  wave  cannot  carry  an  extrinsic  or  orbital  AM  of  the point particle, because   its   position   is undefined. A circularly-polarized electromagnetic plane wave, propagating along the z-axis can be written as

$\vec{E} \space \propto \dfrac{\bar{x} + i\sigma\bar{y}}{\sqrt{2}} \exp{(ikz)},$
$\space \space \space\vec{H} = \bar{z}\vec{E} , $

where $\bar{x}$ , $\bar{y}$ , and $\bar{z}$ denote the unit vectors of the corresponding Cartesian axes, the helicity
parameter $\sigma = ±1$ corresponds to the right-hand and left-hand circular polarizations,
and the wave number $k =ω / c$. The electric (or magnetic) field represents eigenmodes of
the z -component of the spin-1 matrix operator $\hat{S}$ given by:

$\hat{S}= -i\begin{pmatrix}
0 & 0 & 0\\
0 & 0 & 1\\
0 & -1 & 0
\end{pmatrix}\begin{pmatrix}
0 & 0 & -1\\
0 & 0 & 0\\
1 & 0 & 0
\end{pmatrix}\begin{pmatrix}
0 & 1 & 0\\
-1 & 0 & 1\\
0 & 0 & 0
\end{pmatrix},$ $\space \space \space$ $\hat{S_z}E =\sigma E$



Therefore, the plane wave carries the spin AM density S defined as the local expectation
value of the operator $\hat{S}$ with the wavefunction E or H.


$S \propto \sigma \dfrac{\textbf{k}}{k}$


Here we wrote the spin density in a form valid for an arbitrary propagation direction of the plane
wave, $ \textbf{k} / k $ .

# **Spin and polarization in 2D and 3D fields**

**For a 2D paraxial field**

---


For a circularly polarized EM wave, the electric field vector rotates about the axis (which is along the direction of propagation of light). This rotation gives rise to a spin angular momentum (as the electric and magnetic fields carry energy).  
The spin angular density of a general monochromatic electromagnetic field in free space is defined as:

$$\vec{S}=\displaystyle \frac{1}{16\pi\omega}Im(\vec{E}^*\times \vec{E}+\vec{H}^* \times \vec{H}) \equiv S^{(e)}+S^{(m)}$$,

where $S^{(e)}$ is the spin angular momentum density due to electric field and $S^{(m)}$ is the spin angular momentum density due to magnetic field.
The intensity of the field can be determined by the ratio of its energy density to
the frequency: $I=\dfrac{W}{\omega}$

$$I=\displaystyle \frac{1}{16\pi\omega}(\lvert E\rvert ^2+\lvert H\rvert^2) \equiv I^{(e)}+I^{(m)} $$
  

Polarization of a 2D paraxial field
propagating in the positive-z direction is determined by the 2×2 polarization (density) matrix or,
equivalently, by 4 real Stokes parameters, $\vec{s} = (s_0 ,s_1,s_2 ,s_3)$:

$$\hat{\Phi}^{2D(e)} =
\begin{pmatrix}
\langle E_x^*E_x\rangle & \langle E_x^*E_y\rangle\\
\langle E_y^*E_x\rangle & \langle E_y^*E_y\rangle
\end{pmatrix}=\frac{1}{2}
\begin{pmatrix}
s_0^{(e)}+s_1^{(e)} & s_2^{(e)}+\iota s_3^{(e)}\\
s_2^{(e)}-\iota s_3^{(e)} & s_0^{(e)}-s_1^{(e)}
\end{pmatrix}$$,

where <...> denotes time averaging. The above relation shows that stokes parameters are just the coefficients of the expansion of polarization matrix in pauli matrix bases. The corresponding degree of 2D polarization is given by:


$$P^{2D(e)}=\frac{\sqrt{\sum_{i=1}^{3} s_i^{(e)2}}}{s_0^{(e)}}\in[0,1]$$



 By using the definition of $S$, and the relation between the polarization matrix and stokes parameters, it is easy to deduce the following:
 $$\displaystyle \frac{\langle \mathbf{S} ^{(e,m)} \rangle }{\langle I ^{(e,m)} \rangle } = \frac{s_3^{(e,m)}}{s_0^{(e,m)}} \overline{\mathbf{z}}, $$

 where the overbar denotes the unit vector of the corresponding axis. For totally polarized light,
$P^{2D(e,m)} = 1$, the averaging brackets can be omitted, and the usual relation between
the plane-wave spin and degree of circular polarization (helicity). For totally unpolarized
2D field, $P^{2D(e,m)} = 0$ , the Stokes vector is
$\vec{s}^{(e,m)} \propto (1,0,0,0)$ , and the time-averaged spin vanishes:
$\langle S^{(e,m)}\rangle = 0$.

**For a generic 3D nonparaxial field**


---

Polarisation of a generic 3D nonparaxial field at a given point is determined by the
3×3 polarization (density) matrix or, equivalently, by 9 real polarization parameters $\,\,\vec{\Lambda}=(\Lambda_0,\Lambda_1,....,\Lambda_8)$:
$$\hat{\Phi}^{3D(e)} =
\begin{pmatrix}
\langle E_x^*E_x\rangle & \langle E_x^*E_y\rangle & \langle E_x^*E_z\rangle\\
\langle E_y^*E_x\rangle & \langle E_y^*E_y\rangle  & \langle E_y^*E_z\rangle\\
\langle E_z^*E_x\rangle & \langle E_z^*E_y\rangle  & \langle E_z^*E_z\rangle
\end{pmatrix}=\displaystyle \frac{1}{3}
\begin{pmatrix}
\Lambda_0^{(e)}+\Lambda_3^{(e)}+\frac{\Lambda_8^{(e)}}{\sqrt3} & \Lambda_1^{(e)}-\iota\Lambda_2^{(e)} & \Lambda_4^{(e)}-\iota\Lambda_5^{(e)}\\
\Lambda_1^{(e)}+\iota\Lambda_2^{(e)} & \Lambda_0^{(e)}-\Lambda_3^{(e)}+\frac{\Lambda_8^{(e)}}{\sqrt3} & \Lambda_6^{(e)}-\iota\Lambda_7^{(e)}\\
\Lambda_4^{(e)}+\iota\Lambda_5^{(e)} & \Lambda_6^{(e)}+\iota\Lambda_7^{(e)} & \Lambda_0^{(e)}-\frac{2\Lambda_8^{(e)}}{\sqrt3}
\end{pmatrix}$$

The corresponding degree of 3D polarization is given by:


$$P^{3D(e)}=\frac{\sqrt{\sum_{i=1}^{8} \Lambda_i^{(e)2}}}{\sqrt3\Lambda_0^{(e)}}\in[0,1]$$

By using the definition of $S$, and the relation between the polarization matrix and 3D polarization parameters, it is easy to deduce the following:
 $$\displaystyle \frac{\langle \mathbf{S} ^{(e,m)} \rangle }{\langle I ^{(e,m)} \rangle } = -\frac{2\Lambda_7^{(e,m)}}{3\Lambda_0^{(e,m)}} \overline{\mathbf{x}}+\frac{2\Lambda_5^{(e,m)}}{3\Lambda_0^{(e,m)}} \overline{\mathbf{y}}-\frac{2\Lambda_2^{(e,m)}}{3\Lambda_0^{(e,m)}} \overline{\mathbf{z}}$$.

 For totally polarized light, $P^{2D(e,m)}$ = 1, the spin density is still associated with the ellipticity of
the polarization ellipse in 3D space and directed normally to its plane. For completely
unpolarized 3D light (e.g., in the interior of a black-body cavity), $P^{3D(e,m)} = 0$ , the polarization
parameters are
$\vec{\Lambda}^{(e,m)} \propto(1,0,0,...,0)$ , and the time-averaged spin vanishes: $\langle S^{(e,m)} \rangle = 0$.

An important point to keep in mind about the electric and magnetic field
characteristics is that in paraxial light, the electric and magnetic parameters are equal:
$\displaystyle \vec{s}^{(e)} = \vec{s}^{(m)} ≡ \vec{s}, \, \, \, S^{(e)} = S^{(m)} = \frac{S}{2}, \, \, \, \, I^{(e)} = I^{(m)} = \frac{I}{2}$
. This is a consequence of the relation $\vec{H} = \overline{\mathbf{z}} ×\vec{E} $ between the
fields. For nonparaxial light, on the other hand, the electric and magnetic fields are not locally
locked with each other, and therefore in general $\vec{\Lambda}^{(e)} \neq
 \vec{\Lambda}^{(m)} , S^{(e)} \neq S^{(m)} , I^{(e)} \neq I^{(m)}.$



### **Totally unpolarized light in the 2D and 3D sense**

---


Consider a 2D paraxial plane wave with the Electric field oscillating in the transverse directions ($x$ and $y$), the corresponding 2D and 3D matrices look like:

$$\Phi^{2D} =\begin{pmatrix}
1 & 0\\
0 & 1
\end{pmatrix}$$

$$\Phi^{3D} =
\begin{pmatrix}
1 & 0 & 0\\
0 & 1 & 0\\
0 & 0& 0
\end{pmatrix}$$

Now, using the relations between the polarization matrix and the stokes (or their 3D equivalent) parmeters, and the definition of the degree of polarization, one can see that:

$P^{2D(e,m)} = 0$
  
while $ P^{3D(e,m)} = \frac{1}{2}$ 

Thus, a totally unpolarized plane wave is half-polarized in the 3D sense. The spin density vanishes in an unpolarized
plane wave,
$\langle \vec{S}^{(e,m)} \rangle= \vec{0}$ , but its partial polarization in the 3D sense means that the spin can
appear after optical transformations generating a nonzero z-component in the field.

## **Gaussian Beams**

---



A gaussian beam is simply a beam of EM radiation, whose amplitude envelope in the transverse direction is given by the gaussian function. Therefore, the expected form for such a beam, travelling in the z-direction would be:
$\vec{E} = \dfrac{\vec{x}+m\vec{y}}{\sqrt{1+|m|^2}}A(\rho, z)e^{ikz}$
where $A(\rho, z)$ signifies the Gaussian envelope:

$$A(\rho, z)= A_0\dfrac{z_R}{q(z)}e^{ik\frac{\rho^2}{2q(z)}}$$, 

where $q(z)=z-iz_R$ is a complex parameter which characterizes the beam, $z_R = k\omega_0^2/2$ is the Rayleigh diffraction lenght and $\omega_0$ is the beam waist. However, the above mentioned expression does not satisfy the Gauss's law (ironic?), and hence, to satisfy $\nabla \cdot \vec{E}=0$, we need to introduce a longitudinal electric field! The final expression looks like:

$$\vec{E} = \dfrac{\vec{x}+m\vec{y}-\frac{x+my}{q(z)}\vec{z}}{\sqrt{1+|m|^2}}A(\rho, z)e^{ikz}$$

Notice, in the focal plane (z=0), the $\vec{z}$ component beocomes purely imaginary (as $q(z)=-iz_R$), which is equivalent to being phase shifted by $\frac{\pi}{2}$. This gives rise to the transverse spin component, which we will try to analyse further in this report.

---
####**Note: m???**


The complex polarization parameter **m** is related to the real normalised($s_0=1)$ Stokes parameters as:

$$s1=\frac{1-|m|^2}{1+|m|^2},\, s2=\frac{2Re(m)}{1+|m|^2},\, s3=\frac{2Im(m)}{1+|m|^2} $$

###**Spin Angular Momentum density of a Gaussian Beam:**

---


Since our interests are to study the transverse spining of *unpolarized* light, we need to condsider an unpolarized gaussian beam. Such an unpolarized beam can be considered as an incoherent superposition of two beams (S12) with same
intensities and orthogonal polarizations (e.g., with $m=0$ and $m=\infty $). Such incoherent superposition means additive properties of quadratic field quantities, including the spin, intensity , and polarization matrix. Therefore, we directly sum the spin angular momentum density of $m=0$ and $m=\infty $, by substituting the Equation of Electric field in the original definition of spin angular momentum density. This results in the following important relation:

$$\vec{S_x} = \dfrac{2A_0^2y}{z_R}e^{-kr^2/z_R}\vec{x}$$

$$\vec{S_y} = -\dfrac{2A_0^2x}{z_R}e^{-kr^2/z_R}\vec{y}$$

$$\vec{S_z} = 0 $$

Another interesting result to note is that the time-averaged value of logitudinal spin vanishes under the case of an unpolarized non-paraxial wave (just like it did for unpolarized paraxial waves). To see this, one can use the Electric field form given above to conclude that the cohenrency matrix would look like:

$$\Phi^{3D} =
\begin{pmatrix}
1 & 0 & -i\tilde{x}\\
0 & 1 & -i\tilde{y}\\
i\tilde{x} & i\tilde{y}& \tilde{r^2}
\end{pmatrix}$$

where $\tilde{x} = x/z_R$ and $\tilde{y} = y/z_R$

For  the  nonparaxial  3D  field  produced  from  an unpolarized   source, the electric-and   magnetic-field-based   polarization   parameters are equivalent,  similarly to  the  paraxial  case.
 
From $\Phi^{3D}$ we get $\Lambda_1=\Lambda_2=\Lambda_3=\Lambda_4=\Lambda_6=0,$
$$\frac{\Lambda_8}{\Lambda_0} \approx \frac{\sqrt{3}}{2}\frac{1-\tilde{r}^2}{1+\tilde{r}^2/2},\frac{\Lambda_5}{\Lambda_0} \approx \frac{3}{2}\frac{\tilde{x}}{1+\tilde{r}^2/2},\frac{\Lambda_7}{\Lambda_0} \approx \frac{3}{2}\frac{\tilde{y}}{1+\tilde{r}^2/2}$$

From this, 

$$\displaystyle \frac{\langle S\rangle}{\langle I\rangle}=\frac{\langle S^{(e)}\rangle}{\langle I\rangle}=\frac{\langle S^{(m)}\rangle}{\langle I\rangle}\approx\frac{\tilde{x}\bar{y}-\tilde{y}\bar{x}}{1+\tilde{r}^2/2}$$

Thus, depolarization of the incident paraxial wave eliminates the longitudinal spin determined by the  Stokes  parameters,$\frac{S_0}{I_0}=\frac{s_3}{s_0}\bar{z}$,  (because  we  consider  an incoherent superposition  of states with opposite $s_3$), while the transverse spin $S_⊥$remains unaffected.

## **Evanescent waves**
 
---
 
 
This is another form of EM radiations, which carry a longitudinal electric field in them. A general defintion of evanescent waves is an oscillating electric and/or magnetic field that does not propagate as an electromagnetic wave but whose energy is spatially concentrated in the vicinity of the source. But for our purposes, we will stick to the special case of waves which are emitted in total internal reflection. As we have already seen in lectures, during a TIR, some amount of EM field actually enters the optically rarer medium, but the amplitude of it's field decays exponentially. This is  an example of evanescent waves, as the field is practically bounded to be close to the interface (and nearly zero far away perpendicular to it).
To find the expression of electric field associated with it, we begin with a generic form:
 
$$\vec{E} = \vec{A}e^{ik_zz-\kappa x}$$
 
where $\vec{A}$ is the (complex) vector amplitude.
Here, the direction of perpendicular to the interface is in $x$. 
 Again, after imposing maxwell's equations, we get the Electric field as:
 
$$\vec{E} = \dfrac{\vec{x}+m\frac{k}{k_z}\vec{y}-i\frac{\kappa}{k_z}\vec{z}}{\sqrt{1+|m|^2}}e^{ik_zz-\kappa x}$$
 
where $\kappa = \sqrt{k_z^2-k^2}=-\iota k_x$, here $\kappa$ is real.
 
So, as is evident, (taking $m=0$ for simplicity) the phase of $z$ component is again shifted by $\pi/2$. This corresponds to a wave travelling in the $z$ direction, decaying in the $x$ direction, and the $x$ component is $90$ degrees out of phase, which means that the field is circularly polarized in the $x-z$ plane, and moving in the $z$ direction. This is just like the cycloid motion we have seen in earlier classes. Moreover, this gives rise to the transverse spin (remember, circularly polarized light carried spin angular momentum), in the $y$ direction.

evanescent spin.PNG

For simplicity, we can assume that the transmission coefficients of the total internal reflection
are polarization-independent, the generation of the evanescent field can be regarded as a transition from the plane-wave limit $κ = 0 ,  k_z = k$ , to the given finite $κ$ (assuming the local z-axis to be aligned with the propagation direction in both incident and evanescent fields). 

The incident plane wave has the same properties as in the focused-field case, including the normalized spin density, $\vec{S_0} / I_0 = (s_3 / s_0)\vec{z}$ . 

In the evanescent field with finite $κ$ , the normalized
spin density is described by -

$$ \dfrac{\vec{S}}{I} = \dfrac{k}{k_z}\dfrac{\vec{S_0}}{I_0} + \dfrac{\kappa}{k_z}\vec{y} = \dfrac{\vec{S}_\parallel}{I} + \dfrac{\vec{S}_\perp}{I}$$

As before, the evanescent wave produced from a totally unpolarized plane wave can be
described as an incoherent superposition of two waves with orthogonal polarizations (e.g.,
with m = 0 and m = ∞). Direct calculations with fields result in the following polarization
matrix:

$$\Phi^{3D^{(e)}} = \Phi^{3D^{(m)}} \propto
\begin{pmatrix}
1 & 0 & -i\kappa/k_z\\
0 & k^2/k^2_z & 0\\
i\kappa/k_z & 0 & \kappa^2/ k^2_z
\end{pmatrix}$$


As in the case of the focused field, the electric and magnetic field based polarization parameters are equivalent. 


Here we'll have, $\Lambda_1 = Λ_2 = Λ_4 = Λ_6 = Λ_7 = 0$ , and, 

$$ \dfrac{\Lambda_8}{\Lambda_0} = \dfrac{\sqrt{3}}{2}\dfrac{k^2 - \kappa^2/2}{k^2_z}, \space \space \dfrac{\Lambda_3}{\Lambda_0} = \dfrac{3}{4}\dfrac{\kappa^2}{k^2_z}, \space \space \dfrac{\Lambda_5}{\Lambda_0} = \dfrac{3}{2}\dfrac{\kappa}{k_z} $$

From here, the 3D degree of polarization equals, $ P^{3D} = \dfrac{\sqrt{1 + 3\kappa^4/ k^2_z}}{2}$

For $κ / k_z ≪1,$

$$P^{3D} \approx \dfrac{1}{2}(1 + \dfrac{3}{2}\dfrac{\kappa^4}{k^2_z}) $$


Hence, we obtain the time-averaged spin density in the
evanescent wave generated from an unpolarized source:

$$\dfrac{\langle \vec{S} \rangle}{\langle I \rangle} = \dfrac{\langle \vec{S}^{(e)} \rangle}{\langle I^{(e)} \rangle} = \dfrac{\langle \vec{S}^{(m)} \rangle}{\langle I^{(m)} \rangle} = \dfrac{\kappa}{k_z}\vec{y} = \frac{\vec{S}_\perp}{I}$$

As before, depolarization of the incident paraxial wave eliminates the longitudinal spin
determined by the Stokes parameters, while the transverse spin $\vec{S}_⊥$ remains unaffected, and with
equal electric and magnetic field contributions. For polarized evanescent waves, these
contributions depend on the stokes parameters of the incident field.

#####**Stokes Parameter Retrieval**

---
 In  order  to reconstruct the Stokes parameters describing the polarization state of the scattering signal from a single nanoparticle, a set of 12 measurements were recorded. 
 
 The parameters $s_0, s_1,$ and $s_2$ were measured by sending the scattering signal through a linear polarizer and recording the scattering intensity  for  4  angles  of  the  polarizer:  0°  (along  the z-axis),  45°,  90°  and  135°.

The  remaining Stokes  parameter $s_3$ was  measured  by  inserting  a  quarter-waveplate(QWP),  with  its  fast  axis set along the z-axis, just before the linear polarizer and recording the scattering intensity for the linear polarizer set at 45° and 135°. The same set of 6 measurements was repeated on an area of the   sample   without a nanoparticle   (background   measurements),   in   order   to   remove   any contributions from parasite signals to the scattering signal of the nanoparticle.

Define,
$$I^{sc}(\theta)=I^{sc}_p(\theta)-I^{sc}_0(\theta)$$
where,$$ I^{sc}_p= \text{scattering intensities with nanoparticle}$$
$$ I^{sc}_0= \text{scattering intensities without nanoparticle}$$
θ is  the  angle  of  the  linear polarizer.
$$s_0=I^{sc}(0°)+I^{sc}(90°)$$
$$s_1=I^{sc}(0°)-I^{sc}(90°)$$
$$s_2=I^{sc}(45°)-I^{sc}(135°)$$

$$s_3=\frac{1}{cos\delta}[I^{sc}(\phi=0°,45°)-I^{sc}(\phi=0°,135°)+s_2sin\delta]$$
$\delta$ is the   phase   difference   error   of   the QWP, obtained from   the   manufacturer’s specifications and $\phi$ is the QWP fast axis angle.

# **Conclusion**

We have shown that the extension of the 2D polarization states to 3D by the  pure redirection of wavevectors can generate non-zero spin angular momentum in a 2D paraxial light which was completely unpolarized initially. We have provided theoretical calculations and two sets of experimental measurements, first being the transverse spin generated upon tight focusing of unpolarized paraxial light and the second being the generation of an evanescent wave, i.e. the transverse spin generated upon the total internal reflection of unpolarized paraxial light. We used the definition of the spin angular density of a general monochromatic electromagnetic field in free space to conclude that for a totally unpolarized 2D and 3D light, the corresponding degree of 2D and 3D polarization=0, and the time-averaged spin vanishes: $\langle \vec{S}^{(e,m)} \rangle= \vec{0}$. We further showed that a totally unpolarized plane wave is half-polarized in the 3D sense. We then demonstrated the focused beam and evanescent wave experiment and concluded that depolarization of the incident paraxial wave eliminates the longitudinal spin determined by the Stokes parameters, while the transverse spin  $\vec{S}_⊥$  remains unaffected, and with equal electric and magnetic field contributions.
